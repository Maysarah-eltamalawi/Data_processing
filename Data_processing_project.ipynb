```python
def outlier_remove(df):

    Q1 = df.quantile(0.25)
    Q3 = df.quantile(0.75)
    IQR = Q3-Q1

    min_bound = Q1 - 1.5*IQR
    max_bound = Q3 + 1.5*IQR

    df_out = df[~((df < (min_bound)) | (df > (max_bound))).any(axis=1)]
    print("Before: ",df.shape, "After: ", df_out.shape )

    return df_out
```

```python
df_out=outlier_remove(df)
```

```python
from sklearn.preprocessing import OrdinalEncoder

encoder = OrdinalEncoder()
df_encoded = df_out.copy()  # Create a copy of the DataFrame

# Get the list of categorical columns
cat_columns = [col for col in df_encoded.columns if df_encoded[col].dtype == 'object']

# Apply ordinal encoding to categorical columns
df_encoded[cat_columns] = encoder.fit_transform(df_encoded[cat_columns])

# Print the encoded DataFrame
print(df_encoded)
```

```python
df_encoded.drop('accident_year',axis=1,inplace=True)
df_encoded.drop('accident_index',axis=1,inplace=True)
```

```python
from sklearn.model_selection import train_test_split

X = df_encoded.drop('accident_severity', axis=1)  # Independent variables
y = df_encoded['accident_severity']  # Dependent variable
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import mean_squared_error

# Split data into training, validation, and test sets
X_train_val, X_test, y_train_val, y_test = train_test_split(X, y, test_size=0.2)
X_train, X_val, y_train, y_val = train_test_split(X_train_val, y_train_val, test_size=0.25)

# Set range of degrees to test
degrees = range(1, 11)

# Initialize lists to store errors
train_errors = []
val_errors = []

# Loop over degrees
for degree in degrees:
    # Create polynomial features
    poly = PolynomialFeatures(degree)
    X_train_poly = poly.fit_transform(X_train)
    X_val_poly = poly.transform(X_val)

    # Fit Ridge regression model
    model = LogisticRegression()
    model.fit(X_train_poly, y_train)

    # Make predictions
    y_train_pred = model.predict(X_train_poly)
    y_val_pred = model.predict(X_val_poly)

    # Calculate mean squared errors
    train_error = mean_squared_error(y_train, y_train_pred)
    val_error = mean_squared_error(y_val, y_val_pred)

    # Append errors to lists
    train_errors.append(train_error)
    val_errors.append(val_error)

# Find degree with lowest validation error
best_degree = degrees[val_errors.index(min(val_errors))]

# Create polynomial features with best degree
poly = PolynomialFeatures(best_degree)
X_train_val_poly = poly.fit_transform(X_train_val)
X_test_poly = poly.transform(X_test)

# Fit Ridge regression model with best degree
model = LogisticRegression()
model.fit(X_train_val_poly, y_train_val)

# Make predictions on test set
y_test_pred = model.predict(X_test_poly)

# Calculate generalization error (mean squared error on test set)
generalization_error = mean_squared_error(y_test, y_test_pred)
```
